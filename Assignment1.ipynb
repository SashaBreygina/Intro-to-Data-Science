```python
data = pd.read_csv("Automobile_data.csv")
```

#Data Wrangling

```python
data.replace("?", np.NaN, inplace = True)
```

##Finding the Missing Data

```python
#Checking for anymore missing data
data.isnull().sum()

#Using a for loop
for column in data.columns:
  print(column)
  print(data[column].isnull().sum())
```

##Handling Missing Data


```python
#Delete all rows where price is missing
data.dropna(subset = ["price"], inplace = True, axis = 0)

#Reset the index
data.reset_index(drop=True, inplace= True)
```

```python
#Replacing empty normalized losses with mean
# Convert non-numeric values to NaN in the "normalized-losses" column
data["normalized-losses"] = pd.to_numeric(data["normalized-losses"], errors='coerce')

# Calculate the mean of the "normalized-losses" column, excluding NaN values
average_normalized_loss = data["normalized-losses"].mean()

# Replace NaN values in the "normalized-losses" column with the calculated mean
data["normalized-losses"].fillna(average_normalized_loss, inplace=True)
```

```python
#Replace empty stroke with mean
data["stroke"] = pd.to_numeric(data["stroke"], errors='coerce')

average_stroke = data["stroke"].mean()

data["stroke"].fillna(average_stroke, inplace=True)
```

```python
#Replacing null values in bore column with the mean
#average_bore = data["bore"].astype("float").mean(axis = 0)
#data["bore"].replace(np.nan, average_bore, inplace = True)

data["bore"] = pd.to_numeric(data["bore"], errors='coerce')

average_bore = data["bore"].mean()

data["bore"].fillna(average_bore, inplace=True)
```

```python
#Replacing null values in horsepower with mean
#average_horsepower = data["horsepower"].astype("float").mean()
#data["horsepower"].replace(np.nan, average_horsepower, inplace = True)
data["horsepower"] = pd.to_numeric(data["horsepower"], errors='coerce')

average_horsepower = data["horsepower"].mean()

data["horsepower"].fillna(average_horsepower, inplace=True)
```

```python
#Replacing null values in peak-rpm with mean
#average_horsepower = data["peak-rpm"].astype("float").mean()
#data["peak-rpm"].replace(np.nan, average_horsepower, inplace = True)

data["peak-rpm"] = pd.to_numeric(data["peak-rpm"], errors='coerce')

average_horsepower = data["peak-rpm"].mean()

data["peak-rpm"].fillna(average_horsepower, inplace=True)
```

```python
data["price"] = pd.to_numeric(data["price"], errors='coerce')

average_price = data["price"].mean()

data["price"].fillna(average_price, inplace=True)
```

```python
#Replacing null values in num-of-doors with most frequent num-of-do
most_frequent_numofdoor = data["num-of-doors"].value_counts().idxmax()
data["num-of-doors"].replace(np.nan, most_frequent_numofdoor, inplace=True)
```

```python
#Check whether there are any null values after handling the missing data
data.isnull().sum().sum()
```

##Correct Data Format

```python
#Listing the data types for each column
data.dtypes
```

```python
#Converting the data types to the correct format
data[['bore','stroke','price','peak-rpm']] = data[['bore','stroke','price','peak-rpm']].astype('float')
data['normalized-losses'] = data['normalized-losses'].astype('int')
data.dtypes
```

##Data Standardization

```python
#Standardizing the fuel consumption column(city-mpg) from mpg to L/100km
# Convert the "city-mpg" column to numeric (float)
data["city-mpg"] = pd.to_numeric(data["city-mpg"], errors='coerce')

# Perform the division
data["city-L/100km"] = 235 / data["city-mpg"]
```

```python
#Standardizing the column, highway-mpg to highway-L/100km
# Convert the "highway-mpg" column to numeric (float)
data["highway-mpg"] = pd.to_numeric(data["highway-mpg"], errors='coerce')

# Perform the division
data["highway-L/100km"] = 235 / data["highway-mpg"]
```

##Data Normalization

```python
#Normalizing width, height, and length
# Convert columns "length," "height," and "width" to numeric (float)
data["length"] = pd.to_numeric(data["length"], errors='coerce')
data["height"] = pd.to_numeric(data["height"], errors='coerce')
data["width"] = pd.to_numeric(data["width"], errors='coerce')

# Check for NaN or missing values after conversion (if any)
print(data.isnull().sum())

# Perform the division to normalize the columns
data["length"] = data["length"] / data["length"].max()
data["height"] = data["height"] / data["height"].max()
data["width"] = data["width"] / data["width"].max()
```

##Binning

```python
#Convert the data to the correct format
data["horsepower"] = data["horsepower"].astype("float", copy=True)
```

```python
#Visualizing horsepower distribution
plt.hist(data["horsepower"])

plt.xlabel("Horsepower")
plt.ylabel("Amount")
plt.title("Horsepower Distribution")
```

```python
#Constructing the bins
bins = np.linspace(min(data["horsepower"]), max(data["horsepower"]), 4)
bins
```

```python
#Setting the group names
g_name = ["low", "medium", "high"]
```

```python
data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels=g_name, include_lowest = True)
data[['horsepower', 'horsepower-binned']].head(5)
```

```python
#Number fo vehicles in each bin
data["horsepower-binned"].value_counts()
```

```python
#Plot the distribution of each bin
plt.bar(g_name, data["horsepower-binned"].value_counts())
plt.xlabel("Bins")
plt.ylabel("Number of Cars")
plt.title("Bin Vizualization")
```
